```
def printEven(n):
  cnt=0;
  sum=0;
  while(cnt!=n):
    if(cnt%2==0):
      sum=sum+cnt;
    cnt=cnt+1;
  return sum;  
print(printEven(20));    
```

##list with range() function.


```
def linearExample(a):
  sum=0
  for i in range(len(a)):
    if a[i]%3==0 and a[i]%5==0:
      sum=sum+a[i]
  return sum 
a=[1,2,3,4,5,15]
print(linearExample(a))
```

```
def linearFormatedoutput(a): 
  for i in range(len(a)):
      if i==0 or (i==len(a)-1):
          print(a[i])
      elif a[i-1]%2==0 and a[i+1]%2==0:
          print(a[i])
a=[1,6,9,4,16,19,22]        
linearFormatedoutput(a) 
  
```

##Number To LIst Conversion


```
def numberToListConversion(n):
  lst=[]
  while n!=0:
    r=n%10
    lst.append(r)
    n=n//10
  lst.reverse()
  print(lst)
  
n=int(input("enter a number"))
numberToListConversion(n)
```

```
def bubblesort(a):
  for i in range(len(a)-1):
    for j in range(len(a)-1):
      if a[j]>a[j+1]:
        a[j],a[j+1]=a[j+1],a[j]
  for i in range(len(a)):
     print(a[i],end=" ")
lst1=[19,1,25,6,18,3]
bubblesort(lst1)
```
